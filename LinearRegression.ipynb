```python
a=[[1,2],[1,4],[3,9]]
```

```python
x=[]
y=[]
for i in a:
    x.append(i[0])
    y.append(i[1])
avar=[i*0.1 for i in range(-100,101)]
bvar=[i for i in range(-100,101)]
```

```python
def yapprox(xvar,a,b):
    return a*xvar+b
```

```python
def linearReg(xval,yval):
    error=float('inf')

    for avalue in avar:
        for bvalue in bvar:
            inderror=0
            for i in range(len(x)):
                inderror+=(yval[i]-yapprox(xval[i],avalue,bvalue))**2
            if inderror<error:
                error=inderror
                slope=avalue
                intercept=bvalue
                print(slope,intercept)
    return slope,intercept
            
```

```python
slope, intercept=linearReg(x,y)
```
